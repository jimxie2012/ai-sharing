```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import datetime
from warnings import filterwarnings
plt.style.use({'figure.figsize':(12, 6)})
sns.set(font_scale=1.5) 
pd.set_option('display.float_format', lambda x: '%.3f' % x)
plt.rcParams['font.sans-serif']=['SimHei']
```

```python
df = pd.read_csv("./data/test.csv")
data = {"ts":[],"count":[],"time":[]}
for ts,count,raw in zip(df['ts'],df['count'],df['raw']):
    tmp = ts.replace(".000+0000","")
    pos = raw.find(" INFO")
    dd = datetime.datetime.strptime(raw[0:pos].strip(), "%y/%m/%d %H:%M:%S")
    data['ts'].append(tmp)
    data['time'].append(dd)
    data['count'].append(count)
df = pd.DataFrame(data)
df = df.sort_values(by=['time'])
df.reset_index(inplace=True)
df['weekday'] = df['time'].dt.weekday
df['monthday'] = df['time'].dt.month
df['yearday'] = df['time'].dt.year
df['hour'] = df['time'].dt.hour
df['minute'] = df['time'].dt.minute
df['day'] = df['time'].dt.day
def new_column(row):
    return "%02d-%02d"%(row['monthday'],row['day'])
df['test'] = df.apply(lambda row:new_column(row), axis=1)
df = df.sort_values(by=['time'])
df.reset_index(drop=True,inplace=True)
```

```python
df1=df.groupby("test").mean()
df1
```
