```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import datetime
from warnings import filterwarnings
plt.style.use({'figure.figsize':(12, 6)})
sns.set(font_scale=1.5) 
pd.set_option('display.float_format', lambda x: '%.3f' % x)
plt.rcParams['font.sans-serif']=['SimHei']
```

# 读取数据

```python
df = pd.read_csv("./BoxStatus/JP/all.csv")
df.drop(columns=['Unnamed: 0'],inplace=True)
df1 = df[(df['MSP']!='None') & (df['Status']!='unregistered')]
```

# 构造样本

```python
group = df1.groupby('dt')['Status'].count()
df2 = pd.DataFrame()
df2['dt'] = group.index
df2['count'] = group.values
df2['ts'] = pd.to_datetime(df2['dt'])
df2['weekday'] = df2['ts'].dt.weekday
df2['month'] = df2['ts'].dt.month
```

# 变化趋势

```python
df2 = df2.sort_values("ts")
sns.lineplot(x="ts", y="count",data=df2).set_title("整体趋势图")
```

```python
df3 = df2.tail(8)
sns.lineplot(x="ts", y="count",data=df3).set_title("最近一周趋势图")
```

```python
df4 = df2.rolling(window=30,min_periods=1).mean()
sns.lineplot(x=df4.index, y="count",data=df4).set_title("平滑后整体趋势图")
```

```python
df5 = df4.tail(10)
sns.lineplot(x=df5.index, y="count",data=df5).set_title("平滑后最近一周趋势图")
```

# ARIMA分析

```python
from statsmodels.tsa.arima_model import ARMA
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
def Get_error(train_data,pred):
    return (np.sqrt(sum((pred-train_data).dropna()**2/train_data.size)))
```

```python
def Train_once(data,p,d,q):
    train_data = data.dropna()
    train_data = train_data.astype("float64")
    try:
        model = ARIMA(train_data, order=(p,d,q)).fit(disp=-1, method='css')
        predicted = model.predict()
        return model,predicted
    except:
        return None,None
```

```python
def Train_all(data):
    min_error = 65535*65535
    best_model = None
    best_pred = None
    best_p,best_q,best_d = None,None,None
    for p in range(0,5):
        for q in range(0,5):
            for d in range (0,1):
                model,pred = Train_once(data,p,d,q)
                if not model:
                    continue
                error = Get_error(data,pred)
                if error < min_error:
                    min_error = error
                    best_model = model
                    best_pred = pred
                    best_p = p
                    best_q = q
                    best_d = d
    return best_model,best_pred,best_p,best_q,best_d
```

```python
def show_result(train,pred,p,q,d):
    plt.figure(facecolor='white')
    pred.plot(color='blue', label='Predict')
    train.plot(color='red', label='Original')
    error = Get_error(train,pred)
    plt.legend(loc='best')
    plt.title('RMSE: %.4f(%d,%d,%d)'%(error,p,q,d))
    plt.show()
```

# 直接预测

```python
raw_data = df2['count'].tail(30)
model,pred,p,q,d = Train_all(raw_data)
show_result(raw_data,pred,p,q,d)
```

# 分解后预测

```python
raw_data = df4['count'].tail(30)
decomposition = seasonal_decompose(raw_data, period=7,model="additive")
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
trend.dropna(inplace=True)
seasonal.dropna(inplace=True)
residual.dropna(inplace=True)
```

```python
raw_data.plot().set_title("原始趋势")
plt.show()
trend.plot().set_title("趋势分量")
plt.show()
seasonal.plot().set_title("季节分量")
plt.show()
residual.plot().set_title("余项分量")
plt.show()
```

# 分解后预测

```python
model,pred,p,q,d = Train_all(trend)
show_result(trend,pred,p,q,d)
```

```python
model,pred,p,q,d = Train_all(seasonal)
show_result(seasonal,pred,p,q,d)
```

```python
model,pred,p,q,d = Train_all(residual)
show_result(residual,pred,p,q,d)
```
