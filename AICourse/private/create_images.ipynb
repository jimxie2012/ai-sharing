```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
cap = cv2.VideoCapture("./images/da_feng_che.mp4")
images_list = []
for i in range(10):
    ret, frame = cap.read()
    if i % 5 == 0:
        img = frame[230:450,180:375]
        #img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        images_list.append(img)
f,ax = plt.subplots(1,len(images_list), figsize=(5,5))
for i in range(len(images_list)):
    ax[i].set_title(str(i*5))
    ax[i].imshow(images_list[i])
    ax[i].axis('off')
    cv2.imwrite("./images/da_feng_che_"+str(i)+".jpg",images_list[i])
```

计算光流(稠密光流)

```python
img_0=cv2.imread("./images/da_feng_che_0.jpg") #读取第一帧的图像
img_1=cv2.imread("./images/da_feng_che_1.jpg") #读取第二帧的图像
img_0 = cv2.cvtColor(img_0, cv2.COLOR_BGR2RGB) #转换为RGB格式
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB) #转换为RGB格式
gray_0 = cv2.cvtColor(img_0.copy(), cv2.COLOR_RGB2GRAY) #转换为灰度图
gray_1 = cv2.cvtColor(img_1.copy(), cv2.COLOR_RGB2GRAY) #转换为灰度图

flow = cv2.calcOpticalFlowFarneback(gray_0,gray_1, None, 0.5,1,5,1,5,1.25, 0)
mag, ang = cv2.cartToPolar(flow[...,0], flow[...,1])
hsv = np.zeros_like(img_0)
hsv[...,0] = ang*180/np.pi/2
hsv[...,2] = cv2.normalize(mag,None,0,255,cv2.NORM_MINMAX)
rgb = cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)
plt.imshow(rgb)
```
