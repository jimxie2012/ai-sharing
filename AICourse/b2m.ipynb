```python
import numpy, os, array
from scipy import misc
import matplotlib.pyplot as plt
```

```python
filename = './scripts/victim.txt'
graphname = './b2m.png'
 
f = open(filename,'rb') # 
ln = os.path.getsize(filename) # byte
width = 256 # 256
height = 256 #int(len(a)/width)
rem = ln%width #  
a = array.array("B") # uint8 
a.fromfile(f,ln-rem) # a
f.close()
a = a[0:height*width]
g = numpy.reshape(a,(height,width)) # 
g = numpy.uint8(g)
misc.imsave(graphname,g)
plt.imshow(g)
```
