```python
import numpy, os, array
from scipy import misc
import matplotlib.pyplot as plt
```

```python
filename = './scripts/victim.txt'
graphname = './b2m.png'
 
f = open(filename,'rb') # 读入文件
ln = os.path.getsize(filename) # 文件长度（byte）
width = 256 # 固定图片宽度为256
height = 256 #int(len(a)/width)
rem = ln%width # 计算余出的字节 
a = array.array("B") # uint8 数组
a.fromfile(f,ln-rem) # 将文件读入数组a中，舍去余出的字节
f.close()
a = a[0:height*width]
g = numpy.reshape(a,(height,width)) # 将数组转为二维
g = numpy.uint8(g)
misc.imsave(graphname,g)
plt.imshow(g)
```
